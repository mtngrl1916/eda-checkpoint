<h3>Compare Stations

```python
# Plot min and max temp and precipitation by station for common dates to compare data
style.use('ggplot')
plt.rcParams['figure.figsize'] = (20,10)
Compare_Stations = weather_df[(weather_df.DATE>=Station1['DATE'].min().strftime("%B %d %Y")) & (weather_df.DATE<=Station1['DATE'].max().strftime("%B %d %Y"))]
Compare_Stations['STATION'] = Compare_Stations['STATION'].replace(['US1MOCY0003','USW00013988'],['NKC','Downtown'])
Compare_Stations['DATE'] = Compare_Stations['DATE'].astype(str)
Stations_MaxTemp = pd.pivot_table(Compare_Stations, index='STATION',columns='DATE',values='Highest Hourly Temp (Fahrenheit)',aggfunc=np.mean)
Stations_MaxTemp

```

```python
# No NKC data in pivot table.  Is there any data in the necessary columns for NKC?

NKC_Test = Compare_Stations[Compare_Stations['STATION']=='NKC']

columns_to_compare = ['Highest Hourly Temp (Fahrenheit)','Lowest Hourly Temp (Fahrenheit)','Precipitation (inches)','Avg Hourly Temp (Fahrenheit)','Temperature at Observation (Fahrenheit)']

for column in columns_to_compare:
    print(column)
    print(NKC_Test[column].value_counts())
    print("")
```

```python
# No temperature data at NKC Station.  The only main paramter in common is precipitation.  Graphing this to compare.
Compare_Stations['DATE'] = Compare_Stations['DATE'].astype(str)
Stations_Precipitation = pd.pivot_table(Compare_Stations, index='STATION',columns='DATE',values='Precipitation (inches)',aggfunc=np.mean)

fig1 = Stations_Precipitation.loc['Downtown',:].plot(kind = 'line', color = 'r')
Stations_Precipitation.loc['NKC',:].plot(kind = 'line', color = 'b')
plt.title('Precipitation Reported 1998 - 2018',color = 'black')
plt.xlabel('Dates',color = 'black')
plt.ylabel('Inches of Precipitation',color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')

plt.show()
```

```python
# There is too much data in the graph to meaningfully compare the stations. 

# Eliminate dates with null values for NKC station.

dates_to_compare = Stations_Precipitation.columns[Stations_Precipitation.loc['NKC'].notnull()]

fig2 = Stations_Precipitation.loc['Downtown',dates_to_compare].plot(kind = 'line', color = 'r')
Stations_Precipitation.loc['NKC',dates_to_compare].plot(kind = 'line', color = 'b')
plt.title('Precipitation Reported 2006 - 2018',color = 'black')
plt.xlabel('Dates',color = 'black')
plt.ylabel('Inches of Precipitation',color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')

plt.show()
```

```python
# Still too much data by date. Plot totals by month

Stations_Precipitation_Rev = pd.pivot_table(Compare_Stations, index='STATION',columns='Year & Month',values='Precipitation (inches)',aggfunc=np.sum)
Stations_Precipitation_Rev

months_to_compare = Stations_Precipitation_Rev.columns[Stations_Precipitation_Rev.loc['NKC'].notnull()]

fig3 = Stations_Precipitation_Rev.loc['Downtown',months_to_compare].plot(kind = 'area', color = 'r')
Stations_Precipitation_Rev.loc['NKC',months_to_compare].plot(kind = 'area', color = 'b')
plt.title('Precipitation Reported 2006 - 2018',color = 'black')
plt.xlabel('Dates',color = 'black')
plt.ylabel('Inches of Precipitation',color = 'black')
plt.legend(fontsize = 20,labelcolor='k')
plt.xticks(color = 'black')
plt.yticks(color = 'black')

plt.show()
```

<h3>Max and Min Temperature by Year

```python
# I want to see what the temperature variations look like year by year.  I will use a box plot to view this.

Years = weather_df.Year[weather_df.STATION == 'USW00013988'].unique()

weather_df_downtown = weather_df[weather_df.STATION == 'USW00013988']
weather_df_downtown.set_index('Month & Day', inplace=True)

Max_Temp = pd.pivot_table(weather_df_downtown, index='Month & Day',columns='Year',values='Highest Hourly Temp (Fahrenheit)',aggfunc=np.mean)

Max_Temp.plot(kind = 'box', vert=False)
```

```python
Min_Temp = pd.pivot_table(weather_df_downtown, index='Month & Day',columns='Year',values='Lowest Hourly Temp (Fahrenheit)',aggfunc=np.mean)

Min_Temp.plot(kind = 'box', color='b', vert=False)
```

<h3> Average Max and Min Temp by Day

```python
# Line chart for average max and min temp by day

Avg_MaxMin = pd.pivot_table(weather_df_downtown, index='Month & Day',values=['Highest Hourly Temp (Fahrenheit)','Lowest Hourly Temp (Fahrenheit)'],aggfunc=np.mean)

ax = Avg_MaxMin.plot(kind='line')
plt.axhline(y = 32, color = 'k', linestyle = ':',label="Freezing")
plt.title('Average Max and Min Temperature by Day',color = 'black')
plt.xlabel('Month-Day of the Year',color = 'black')
plt.ylabel('Temperature (Fahrenheit)',color = 'black')
plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper center')
plt.show()
```

<h3>Last Spring Frost Date

```python
weather_df_downtown.reset_index(inplace = True)
```

```python
# Plot last Spring frost date by year

First_Half = weather_df_downtown[(weather_df_downtown['Month'] < 7) & (weather_df_downtown['Lowest Hourly Temp (Fahrenheit)'] <= 32)]
Last_Frost = pd.pivot_table(First_Half, index = ['Year'],values=['Month & Day'],aggfunc=np.max)
Last_Frost.reset_index(inplace = True)

order = sorted(Last_Frost['Month & Day'].unique())
Last_Frost['Month & Day'] = pd.Categorical(Last_Frost['Month & Day'],order)
Last_Frost.sort_values(by=['Month & Day'],inplace=True)
plt.scatter(Last_Frost['Year'],Last_Frost['Month & Day'])
```

<h3>First Fall Frost Date

```python
# Plot first Fall frost date by year
Second_Half = weather_df_downtown[(weather_df_downtown['Month'] > 6) & (weather_df_downtown['Lowest Hourly Temp (Fahrenheit)'] <= 32)]

First_Frost = pd.pivot_table(Second_Half, index = ['Year'],values=['Month & Day'],aggfunc=np.min)
First_Frost.reset_index(inplace=True)

order = sorted(First_Frost['Month & Day'].unique())
First_Frost['Month & Day'] = pd.Categorical(First_Frost['Month & Day'],order)
First_Frost.sort_values(by=['Month & Day'],inplace=True)
plt.scatter(First_Frost['Year'],First_Frost['Month & Day'])
```

<h3>Average Precipitation by Day

```python
# Plot annual average precipitation by day

Avg_Precip = pd.pivot_table(weather_df_downtown, index='Month & Day',values='Precipitation (inches)',aggfunc=np.mean)
Avg_Precip
ax = Avg_Precip.plot(kind='line',color='b')
plt.title('Average Precipitation by Day',color = 'black')
plt.xlabel('Month-Day of the Year',color = 'black')
plt.ylabel('Precipitation (inches)',color = 'black')
plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper center')
plt.show()
```
