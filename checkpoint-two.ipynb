```python
# Review general values present in each column

for column in weather_df.columns:
    print (weather_df[column].value_counts())
    print("")
```

```python
# Correct datatype of Date

weather_df['DATE'] = pd.to_datetime(weather_df['DATE'])
weather_df['DATE']
```

```python
# Isolate day, month, year

weather_df['Year'] = weather_df['DATE'].dt.year 
weather_df['Month'] = weather_df['DATE'].dt.month
weather_df['Day'] = weather_df['DATE'].dt.day

# Binning
weather_df['Month & Day'] = weather_df['Month'].astype(str).apply(lambda x: x.zfill(2)) + '-' + weather_df['Day'].astype(str).apply(lambda x: x.zfill(2))
weather_df['Year & Month'] = weather_df['Year'].astype(str) + '-' + weather_df['Month'].astype(str).apply(lambda x: x.zfill(2))
weather_df.head()
```

```python
# Find all time low temperature

weather_df['Lowest Hourly Temp (Fahrenheit)'].min()
```

<h3>Compare Stations

```python
# Plot min and max temp and precipitation by station for common dates to compare data
style.use('ggplot')
plt.rcParams['figure.figsize'] = (20,10)
Compare_Stations = weather_df[(weather_df.DATE>=Station1['DATE'].min().strftime("%B %d %Y")) & (weather_df.DATE<=Station1['DATE'].max().strftime("%B %d %Y"))]
Compare_Stations['STATION'] = Compare_Stations['STATION'].replace(['US1MOCY0003','USW00013988'],['NKC','Downtown'])
Compare_Stations['DATE'] = Compare_Stations['DATE'].astype(str)
Stations_MaxTemp = pd.pivot_table(Compare_Stations, index='STATION',columns='DATE',values='Highest Hourly Temp (Fahrenheit)',aggfunc=np.mean)
Stations_MaxTemp

```

```python
# No NKC data in pivot table.  Is there any data in the necessary columns for NKC?

NKC_Test = Compare_Stations[Compare_Stations['STATION']=='NKC']

columns_to_compare = ['Highest Hourly Temp (Fahrenheit)','Lowest Hourly Temp (Fahrenheit)','Precipitation (inches)','Avg Hourly Temp (Fahrenheit)','Temperature at Observation (Fahrenheit)']

for column in columns_to_compare:
    print(column)
    print(NKC_Test[column].value_counts())
    print("")
```

```python
# No temperature data at NKC Station.  The only main paramter in common is precipitation.  Graphing this to compare.
Compare_Stations['DATE'] = Compare_Stations['DATE'].astype(str)
Stations_Precipitation = pd.pivot_table(Compare_Stations, index='STATION',columns='DATE',values='Precipitation (inches)',aggfunc=np.mean)

fig1 = Stations_Precipitation.loc['Downtown',:].plot(kind = 'line', color = 'r')
Stations_Precipitation.loc['NKC',:].plot(kind = 'line', color = 'b')
plt.title('Precipitation Reported 1998 - 2018',color = 'black')
plt.xlabel('Dates',color = 'black')
plt.ylabel('Inches of Precipitation',color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')

plt.show()
```

```python
# There is too much data in the graph to meaningfully compare the stations. 

# Eliminate dates with null values for NKC station.

dates_to_compare = Stations_Precipitation.columns[Stations_Precipitation.loc['NKC'].notnull()]

fig2 = Stations_Precipitation.loc['Downtown',dates_to_compare].plot(kind = 'line', color = 'r')
Stations_Precipitation.loc['NKC',dates_to_compare].plot(kind = 'line', color = 'b')
plt.title('Precipitation Reported 2006 - 2018',color = 'black')
plt.xlabel('Dates',color = 'black')
plt.ylabel('Inches of Precipitation',color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')

plt.show()
```

```python
# Plot totals by month

Stations_Precipitation_Rev = pd.pivot_table(Compare_Stations, index='STATION',columns='Year & Month',values='Precipitation (inches)',aggfunc=np.sum)
Stations_Precipitation_Rev

months_to_compare = Stations_Precipitation_Rev.columns[Stations_Precipitation_Rev.loc['NKC'].notnull()]

fig3 = Stations_Precipitation_Rev.loc['Downtown',months_to_compare].plot(kind = 'area', color = 'r')
Stations_Precipitation_Rev.loc['NKC',months_to_compare].plot(kind = 'area', color = 'b')
plt.title('Precipitation Reported 2006 - 2018',color = 'black')
plt.xlabel('Dates',color = 'black')
plt.ylabel('Inches of Precipitation',color = 'black')
plt.legend(fontsize = 20,labelcolor='k')
plt.xticks(color = 'black')
plt.yticks(color = 'black')

plt.show()
```
